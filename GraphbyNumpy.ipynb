```
import numpy as np
import matplotlib.pyplot as plt
a=[1,2,0.5,4]
b=[-2,3,0,1]
plt.scatter(a,b,100)
plt.xlabel("a")
plt.ylabel("b")
plt.title("Follow a and b")
plt.show()
x=np.linspace(-5,5,10)
y1=x-2
y2=2*x+3
y3=0.5*x
y4=4*x+1
plt.plot(x,y1,y3,y4,color="red")
plt.xlabel("x")
plt.ylabel("y")
# dat ten cho graph
plt.title("Follow x and y")
plt.show()
```
